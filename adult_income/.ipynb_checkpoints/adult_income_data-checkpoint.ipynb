## Exporatory data analysis Adult Income dataset

```python
import pandas as pd
import numpy as np
```

```python
# Read data from csv file
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'educational-num','marital-status', 
                'occupation', 'relationship', 'race', 'gender','capital-gain', 'capital-loss', 'hours-per-week', 'native-country','income']
# Train data
train_data  = pd.read_csv('data/train_data.txt', names = column_names)
train_data.head()
```

```python
# Displaying the data statistics
train_data.describe()
```

```python
# Data types of the columns in the dataset
train_data.dtypes
```

```python
# Reading test data
test_data = pd.read_csv('data/test_file.txt', names = column_names)
test_data.head()
```

```python
test_data.describe()
```

```python
test_data.dtypes
```

```python
test_data['income'].replace(regex = True, inplace = True, to_replace = r'\.', value = r'')
```

```python
total_data = pd.concat([train_data, test_data])
total_data.head()
```

```python
total_data.info()
```

```python
total_data.reset_index(drop = True)
```

```python
# Replace the data types of categorical variables to type category
for i in set(total_data.columns) - set(total_data.describe().columns):
    total_data[i] = total_data[i].astype('category')
```

```python
total_data.dtypes
```

```python
# Checking the data for null values
for i, j in zip(total_data.columns, (total_data.values.astype(str) == '?'). axis = 0):
    for j > 0:
        print('{} = {}'.format(i, j))
```
